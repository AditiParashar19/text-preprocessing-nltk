# NLP Based Chatbot - Intent Classifier

```python
#tokenization
from nltk.tokenize import word_tokenize
#stopwords List
from nltk.corpus import stopwords
#Stemming
from nltk.stem import PorterStemmer
#Lemmatization
from nltk.stem import WordNetLemmatizer
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
#This is training data
training_data = [
    ("hello","greet"),
    ("hi","greet"),               #hello - category greet
     ("hello there","greet"),
     ("hey","greet"),
    ("hi there","greet"),
     ("good morning","greet"),
    ("good evening","greet"),
    ("good afternoon","greet"),
     ("what's the weather","weather"),
    ("how's the weather today","weather"),
    ("is it raining","weather"),
    ("tell me temperature","weather"),
    ("what's the temperature","weather"),
    ("open google","open_web"),
    ("open facebook","weather"),
    ("open youtube","open_web"),
     ("go to google","open_web"),
    ("open linkedin","open_web"),
    ("bye","exit"),
    ("exit","exit"),
    ("good bye","exit")
]
```

```python
# we have two columns : output (x),target (y)- > categories
sentences=[]  # features / x
labels=[]     # target / y

for text,intent in training_data:     #tuple
    sentences.append(text)     
    labels.append(intent)
```

### NLP On text data

```python
def preprocess_text(documents):
    punctuations = string.punctuation
    english_stopwords = stopwords.words("english")
    wnet = WordNetLemmatizer()
    processed_text = []
    for i in documents:
        txt = i.lower()
        tokens=word_tokenize(txt)
        filtered_tokens=[]
        for word in tokens:
            if(word not in english_stopwords):
                filtered_tokens.append(word)
        #print("Filtered tokens : ",filtered_tokens)
        clean_tokens = [word for word in filtered_tokens if word not in punctuations]
        lemmatized_words = []
        for word in clean_tokens:
            lemmatized_words.append(wnet.lemmatize(word,"v"))
        #print("After Lemmatization ",lemmatized_words)
        #print("="*55)
        final_tokens =[]
        for word in lemmatized_words:
            if(word.isalpha()):
                final_tokens.append(word)
        cleaned_text = " ".join(final_tokens)
        processed_text.append(cleaned_text)
    return processed_text
```

```python
sentences[:5]
```

```python
cleaned_data = preprocess_text(sentences)
```

```python
cleaned_data
```

# Apply Tfidf 

```python
vectorizer=TfidfVectorizer()
x = vectorizer.fit_transform(cleaned_data)      #convert text data into numbers ie. vectors        fit,transform
print(x.toarray())
```

# Apply Machine Learning

```python
#Train the model 
model = LogisticRegression()
# x-> clean data (vectorize data), labels -> categories
model.fit(x,labels) 
```

# Test the Model

```python
#user_msg = "hello how are you ?"
user_msg="open chatgpt"
processed = preprocess_text([user_msg])               #same work as training data
user_vector = vectorizer.transform(processed)      #only transform because it is testing 
prediction = model.predict(user_vector)
print("Prediction is : ",prediction)
```
